```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("red_wine.csv")
```

```python
dataset.columns=dataset.columns.str.replace(" ","_")
dataset
```

```python
dataset.head(10)
```

```python
dataset.tail(20)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
mean_fa= dataset.fixed_acidity.mean()
mean_va= dataset.volatile_acidity.mean()
mean_c= dataset.chlorides.mean()
mean_fsd= dataset.free_sulfur_dioxide.mean()
mean_d= dataset.density.mean()
mean_q=dataset.quality.mean()  
```

```python
nData=dataset.copy()

dataset['fixed_acidity']=dataset['fixed_acidity'].fillna(mean_fa)
dataset['volatile_acidity']=dataset['volatile_acidity'].fillna(mean_va)
dataset['chlorides']=dataset['chlorides'].fillna(mean_c)
dataset['free_sulfur_dioxide']=dataset['free_sulfur_dioxide'].fillna(mean_fsd)
dataset['density']=dataset['density'].fillna(mean_d)
dataset['quality']=dataset['quality'].fillna(mean_q)
```

```python
dataset.isnull().sum()
```

```python
dataset.duplicated()
```

```python
dataset.drop_duplicates()
```

```python
nData_statistics=dataset.describe()
dataset.describe()
```

```python
 dataset.corr()
```

# PIE CHART
---

```python
nData_statistics[['citric_acid','residual_sugar','chlorides','total_sulfur_dioxide','sulphates','alcohol']].loc['mean']
```

```python
citric_acid=nData_statistics['citric_acid'].loc['mean']
residual_sugar=nData_statistics['residual_sugar'].loc['mean']
chlorides = nData_statistics['chlorides'].loc['mean']
total_sulfur_dioxide = nData_statistics['total_sulfur_dioxide'].loc['mean']
sulphates = nData_statistics['sulphates'].loc['mean']
alcohol = nData_statistics['alcohol'].loc['mean']

labels=['Citric acid','Residual sugar','Chlorides','Total sulfur dioxide','Sulphates','Alcohol']
explode_values=[0.3,0,0.4,0,0.2,0]

plt.pie([citric_acid,residual_sugar,chlorides,total_sulfur_dioxide,sulphates,alcohol], labels = labels,autopct = '%.2f%%',explode = explode_values)
plt.title('RED WINE COMPONENTS',fontdict = {'fontweight':'bold'})
        
```

#### This pie chart shows the average component percentage of all ingredients present in red wine 
#### Varying the precentages of the ingredients is what results in the different kinds of red wine present today.

# pH Histogram
---

```python
bins =np.arange(2.7,4.1,0.1)
plt.figure(figsize=(7,5))
plt.hist(nData.pH,bins=bins,rwidth=0.6)
plt.xticks(bins)
plt.title('pH Histogram')
plt.xlabel('pH')
plt.ylabel('Red Wine')
plt.show()
```

#### The histogram shows that majority of red wines have their pH in the range of 3.1 - 3.5
#### The minority have thier in the ranges of 0-3.1 and 3.5 -4.0

# Scatter Graph:Total sulfur dioxide VS Free sulfur dioxide
---

```python
%matplotlib inline

plt.figure(figsize=(8,6))
plt.title('Total sulfur dioxide VS Free sulfur dioxide',fontdict={'fontweight':'bold'})
plt.scatter(x='total_sulfur_dioxide',y='free_sulfur_dioxide',data = dataset, c='b',alpha=0.6)
plt.xlabel('Total sulfur dioxide')
plt.ylabel('Free sulfur dioxide')
plt.legend()
```

#### From the scatter plot it is seen that wine with higher free sulfur dioxide content generaly have higher total sulfur dioxide content,
#### as implied by the positive slope
